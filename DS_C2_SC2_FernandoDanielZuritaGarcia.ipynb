Imporación de librerías necesarias

```python
import numpy as np
import pandas as pd
```

Extracción de la información del archivo.

```python
country_vaccinations = pd.read_csv('/content/country_vaccinations.csv')
```

Obtención de los tipos de datos del dataframe obtenido

```python
country_vaccinations.dtypes
```

Conversión de los valores de fecha correspondientes a `datetime64` y verificación de nuevos tipos de datos

```python
country_vaccinations['date'] = pd.to_datetime(country_vaccinations['date'])
country_vaccinations.dtypes
```

Determinación de la cantidad de vacunas aplicadas de cada compañía

```python
country_vaccinations[['vaccines', 'daily_vaccinations']].groupby('vaccines').sum()
```

Obtención de la cantidad de vacunas aplicadas en todo el mundo

```python
country_vaccinations['daily_vaccinations'].sum()
```

Cálculo del promedio de vacunas aplicadas por país

```python
country_vaccinations[['country', 'daily_vaccinations']].groupby('country').mean()
```

Determinación de la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo

```python
country_vaccinations[country_vaccinations['date'] == '2021-01-29']['daily_vaccinations'].sum()
```

Creación de dataframe `conDiferencias`

```python
conDiferencias = pd.DataFrame(country_vaccinations)
conDiferencias['diferencias'] = np.abs(conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw'])
conDiferencias
```

Obtención del periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua

```python
country_vaccinations['date'].max() - country_vaccinations['date'].min()
```

Creación de dataframe `conCantidad`

```python
conCantidad = pd.DataFrame(country_vaccinations)
conCantidad['canVac'] = country_vaccinations['vaccines'].str.split(', ').str.len()
```

Creación de dataframe `antes20`

```python
antes20 = country_vaccinations.loc[country_vaccinations['date'] < '2020-12-20']
antes20
```

Creación de dataframe `pfizer`

```python
pfizer = country_vaccinations.loc[country_vaccinations['vaccines'].str.contains('pfizer', case=False)]
pfizer
```

Almacenado de los dataframes

```python
with pd.ExcelWriter("resultadosReto.xlsx", datetime_format='DD/MM/YYYY') as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes20 .to_excel(writer, sheet_name='antes20', index=False)
    pfizer.to_excel(writer, sheet_name='pfizer', index=False)
```
